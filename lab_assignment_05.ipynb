Question 3 (15 points). In this question, you are required to create visualization for time series data. 

We have collected all the articles from AMIA in the last 30 years. The dataset is avaliable on GitHub: https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv. 

Please use the tool Yake: https://github.com/LIAAD/yake to extract the top N (you can set different N value to get the best visualization results) keywords from the article titles (all the articles), then visualize how the keywords changed over years in one figure. (Split the data by every five years)

```python
# Your python code here:

!pip install yake

import pandas as pd
import matplotlib.pyplot as plt
import yake
arim_df = pd.read_csv("https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv", sep=",")
# we drop the nan values rows in year
arim_df = arim_df[arim_df['year'].isna()!=True]

arim_df['year'] = arim_df['year'].astype('int64')

print(arim_df)
# we retrive the top ten keywords in that year
for year in arim_df['year'].unique():
    print("The most top 10 key words of the year {} :".format(year))
    print()
    max_ngram_size = 1
    language = "en"
    nok_words = 10
    text = " ".join(title for title in arim_df[arim_df['year']==year]['title'].values)
    cust_extract = yake.KeywordExtractor(lan=language, top=nok_words, features=None)
    top_k_words = cust_extract.extract_keywords(text)

    for kywords in top_k_words:
        print(kywords)
    print("******************************************")
    


```
